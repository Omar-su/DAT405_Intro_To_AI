```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import sklearn
#To get access to a dataset
from sklearn import datasets
#Import train_test_split function
from sklearn.model_selection import train_test_split
#Import knearest neighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
from sklearn.metrics import confusion_matrix

# Authors : Omar Sulaiman and Yasmine Fahraoui



#Always check the path to the CSV file
df = pd.read_csv('/Users/omars/OneDrive/Desktop/iris.csv')
print(df)

#Counting the number of elements in each species
print(df["species"].value_counts())

#Plotting the data
#plt.scatter(df.loc[:,"petal_length"], df.loc[:,"sepal_length"])
#plt.title("Scatter plot of Sepal_length VS Petal length")
#plt.xlabel("Petal length [mm]")
#plt.ylabel("Sepal length [mm]")
sns.FacetGrid(df, hue = 'species', height=6).map(plt.scatter, 'petal_length', 'sepal_width').add_legend()

#We convert species names into numbers

flowers_mapping = {'setosa' : 1, 'versicolor': 2, 'virginica':3}
df['species'] = df['species'].map(flowers_mapping)
print(df['species'])


X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width' ]].values
Y = df['species'].values

print (X,Y)
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size = 0.25, random_state = 2)
knn = KNeighborsClassifier(n_neighbors=3) #k=1,3,21

#Train the model using the training sets
knn.fit(x_train, y_train)

#Predict the response for test dataset
y_pred = knn.predict(x_test)
#Construction of the prediction matrix which will calculate the accuracy of the model
cm = confusion_matrix(y_test, y_pred)
print("confusion matrix \n",cm)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
plt.figure(figsize=(9,9))

```

```python
# Visualising the confusion matrix in a better way
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(metrics.accuracy_score(y_test, y_pred))
plt.title(all_sample_title, size = 15);
```
